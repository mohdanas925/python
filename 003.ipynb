```python
# Write a program to tell the number of dogs and chickens when the total no of heads and legs are given


heads = int(input('Enter the head: '))
legs = int(input('Enter the legs: '))

chickens = -(legs - 4*heads)//2

dogs = heads - chickens

print('Number of dogs', dogs)
print('Number of chickens', chickens)
```

```python
# Given H, W, B of milk tank and dimensions of glass h = 3cm, r=1cm
# Find out how many glasses of milk can be obtained from a tank
import math

height = int(input('Enter the height of a tank: '))
length = int(input('Enter the length of a tank: '))
breadth = int(input('Enter the breadth of a tank'))

radius = int(input('Enter the radius of a glass: '))
H = int(input('Enter the height of a glass: '))

tank_vol = length*breadth*height
glass_vol = 3.14*radius*H

result = math.floor(tank_vol / glass_vol)
print('The total no of glasses are: ', result)
```

```python
# Reverse a given integer number
number = int(input('Enter the number: '))

rev = 0
while number > 0:
    rem = number%10
    rev = rev*10 + rem
    number = number//10

print(rev)
```

```python
# Take a input n and find the sum from 1 to n. If the number is divisibly by five, skip the num
# if sum greater than 300 exit
n = int(input('Enter the number: '))
sum = 0
for i in range(1, n+1):
    if i%5 == 0:
        continue

    sum += i
    if sum > 300:
        sum -= i
        break
        
print(sum)
```

```python
# WAP that keeps on accepting inputs from user until user gives 0
# Display sum and average of all the nums

sum = 0
count = 0
while True:
    num2 = int(input('Enter the number: '))
    if num2 == 0:
        break
    sum += num2
    count += 1

print('Sum is', sum)
print('Average is', sum/count)
```

```python
# find all nums divisible by 7 but not a multiple of 5 between 2000 and 3200(both included)
# The numbers obtained should be printed in comma separated sequence in single line
L = []
for i in range(2000, 3201):
    if i% 7 == 0 and i%5 != 0:
        L.append(str(i))
print(",".join(L))
```

```python
# Find all nums between 1000 - 3000 included whose all digits are even
L= []
for i in range(1000, 3001):
    flag = True
    curr = i

    while curr > 0:
        rem = curr%10
        if rem%2 != 0:
            flag = False
            break
        curr = curr//10

    if flag:
        L.append(str(i))
print(",".join(L))
```

```python
pos = [0, 0]
while True:
    s = input('Enter the robot path')
    if s == '!':
        break
    direction = s.split()[0]
    steps = int(s.split()[1])

    if direction == 'UP':
        pos[1] = pos[1] + steps
    elif direction == 'DOWN':
        pos[1] = pos[1] - steps
    elif direction == 'LEFT':
        pos[0] = pos[0] - steps
    elif direction == 'RIGHT':
        pos[0] = pos[0] + steps
    else:
        pass

print(pos)
print((pos[0]**2 + pos[1]**2)**0.5)
```

```python
# check whether a number is prime or not
num = int(input('Enter the number: '))

flag = True
for i in range(2, num):
    if num%i == 0:
        flag = False
        break
if flag:
    print('Prime')
else:
    print('Not Prime')
```
