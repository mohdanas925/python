# File Handling

```python
# Types of data used for I/O:
# 1. Text - '12345' as a sequence of unicode chars
# 2. Binary - 12345 as a sequence of bytes of its binary equivalent

# Hence there are 2 file types to deal with
# Text files - All program files are text files
# Binary Files - Images,music,video,exe files
```

```python
# How File I/O is done in most programming languages
# Open a file
# Read/Write data
# Close the file
```

## Writing to a file

```python
# case 1 - if the file is not present
f = open('sample.txt', 'w')
f.write('Hello World')
f.close()
# since file is closed hence this will not work
f.write('hello')
```

```python
# write multiline strings
f = open('sample1.txt','w')
f.write('hello world')
f.write('\nhow are you?')
f.close()
```

```python
# case 2 - if the file is already present
# The write operation on existing file will replace the old data with new one
f = open('sample.txt','w')
f.write('salman khan')
f.close()
```

```python
# Problem with w mode
# introducing append mode
# a ppend mode will append the data in the existing file
f = open('sample1.txt','a')
f.write('\nI am fine')
f.close()
```

```python
# write lines
# Writes a list (or other iterable) of strings to the file. 
L = ['hello\n','hi\n','how are you\n','I am fine']

f = open('sample.txt','w')
f.writelines(L)
f.close()
```

```python
# reading from files
# -> using read()
# Reads the entire file content as a single string (or bytes in binary mode) if size is not specified. 
# If size is an integer, it reads that number of characters/bytes
f = open('sample.txt','r')
s = f.read()
print(s)
f.close()
```

```python
# reading upto n chars
f = open('sample.txt','r')
s = f.read(10)
print(s)
f.close()
```

```python
# readline() -> to read line by line
# Reads a single line from the file.
f = open('sample.txt','r')
print(f.readline(),end='')
print(f.readline(),end='')
f.close()
```

```python
# readline() -> to read line by line
# Reads all lines of the file and returns them as a list of strings.
f = open('sample.txt','r')
print(f.readlines())
f.close()
```

```python
# reading entire using readline

f = open('sample.txt','r')

while True:
  data = f.readline()

  if data == '':
    break
  else:
    print(data,end='')
      
f.close()
```

### Using Context Manager (With)

```python
# It's a good idea to close a file after usage as it will free up the resources
# If we dont close it, garbage collector would close it
# with keyword closes the file as soon as the usage is over
```

```python
# with
with open('sample1.txt','w') as f:
  f.write('selmon bhai')
```

```python
f.write('Hello')
```

```python
# try f.read() now
with open('sample.txt','r') as f:
  print(f.readline())
```

```python
# moving within a file -> 10 char then 10 char
with open('sample.txt','r') as f:
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
```

```python
# benefit? -> to load a big file in memory
big_L = ['hello world ' for i in range(1000)]

with open('big.txt','w') as f:
  f.writelines(big_L)
```

```python
# loading big files data in chunks. Libraries like pandas load data in chunks
with open('big.txt', 'r') as f:
    chunk_size = 10;

    while len(f.read(chunk_size)) > 0:
        print(f.read(chunk_size), end='***')
        f.read(chunk_size)
```

```python
# seek and tell function
with  open('sample.txt', 'r') as f:
    f.seek(0)
    print(f.read(10))
    print(f.tell())

    f.seek(0)
    print(f.read(10))
    print(f.tell())

    #The seek() function is used to change the position of the file pointer to a specific location within the file. 
    #This allows for random access to different parts of the file for reading or writing

    #The tell() function is used to return the current position of the file pointer within the file. 
    #The value returned is an integer representing the number of bytes from the beginning of the file
```

```python
# seek during write
with open('sample.txt', 'w') as f:
    f.write('Hello')
    f.seek(0)
    f.write('Xa')

with  open('sample.txt', 'r') as f:
    f.seek(0)
    print(f.read(10))
```

## Serialization and Deserialization

```python
# Serialization - process of converting python data types to JSON format
# Deserialization - process of converting JSON to python data types
```

```python
# JSON
# JSON (JavaScript Object Notation) is a lightweight data format used to store and exchange data between systems (especially between server and web applications)

# Human-readable
# Easy to write
# Language-independent
# Widely used in APIs and web services
```

```python
{
  "name": "Anas",
  "skills": ["Python", "SQL", "Machine Learning"],
  "marks": {
    "math": 85,
    "statistics": 90
  }
}
```

```python
# serialization using json module
# list
import json

L = [1,2,3,4]

with open('demo.json','w') as f:
  json.dump(L,f)
```

```python
# dict
d = {
    'name':'nitish',
     'age':33,
     'gender':'male'
}

with open('demo.json','w') as f:
  json.dump(d,f,indent=4)
```

```python
# deserialization
import json

with open('demo.json','r') as f:
  d = json.load(f)
  print(d)
  print(type(d))
```

```python
# serialize and deserialize tuple
# tuple stores in json file as a list
import json

t = (1,2,3,4,5)

with open('demo.json','w') as f:
  json.dump(t,f)
```

```python
# serialize and deserialize a nested dict

d = {
    'student':'nitish',
     'marks':[23,14,34,45,56]
}

with open('demo.json','w') as f:
  json.dump(d,f)
```

## Serializing and Deserializing custom objects

```python
class Person:

  def __init__(self,fname,lname,age,gender):
    self.fname = fname
    self.lname = lname
    self.age = age
    self.gender = gender

# format to printed in
# -> Nitish Singh age -> 33 gender -> male
```

```python
person = Person('Nitish','Singh',33,'male')
```

```python
# As a string
import json

def show_object(person):
  if isinstance(person,Person):
    return "{} {} age -> {} gender -> {}".format(person.fname,person.lname,person.age,person.gender)

with open('demo.json','w') as f:
  json.dump(person,f,default=show_object)
```

```python
# As a dict
import json

def show_object(person):
  if isinstance(person,Person):
    return {'name':person.fname + ' ' + person.lname,'age':person.age,'gender':person.gender}

with open('demo.json','w') as f:
  json.dump(person,f,default=show_object,indent=4)
```

```python
# deserializing
import json

with open('demo.json','r') as f:
  d = json.load(f)
  print(d)
  print(type(d))
```

## Pickling

```python
# Pickling is the process whereby a Python object hierarchy is converted into a byte stream, and unpickling is the inverse operation, 
# whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy.
```

```python
class Person:

  def __init__(self,name,age):
    self.name = name
    self.age = age

  def display_info(self):
    print('Hi my name is',self.name,'and I am ',self.age,'years old')
```

```python
p = Person('nitish',33)
```

```python
# pickle dump
import pickle
with open('person.pkl','wb') as f:
  pickle.dump(p,f)
```

```python
# pickle load
import pickle
with open('person.pkl','rb') as f:
  p = pickle.load(f)

p.display_info()
```

```python
# Pickle Vs Json
# Pickle lets the user to store data in binary format. 
# JSON lets the user store data in a human-readable text format.
```
