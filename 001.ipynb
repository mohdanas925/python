# 3. Variables

```python
# Static vs Dunamic Typing

# Static typing checks variable types at compile-time(before-execution), catching errors early for safer, faster performance but requiring explicit declarations (e.g., Java, C++)
# Variables types must be explicitly declared eg.(int age = 20)

# while dynamic typing checks types at runtime(during-execution), offering more flexibility, faster initial coding, but potentially slower execution and runtime bugs (e.g., Python, JavaScript)
# No need to declare variables types explicitly eg.(age = 20)
```

```python
# Static Binding vs Dynamic Binding

# Dynamic Binding (Data type of variable may change)
# Same variable can hold different values of different data-types like in python
a = 5
print(a)
a = 'Anas'
print(a)


# Static Binding 
# Same variable can hold only different values of one data-type like in c/c++/java
```

```python
a,b,c = 1,2,3
print(a,b,c)
```

```python
a=b=c = 5
print(a,b,c)
```
