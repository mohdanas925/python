# Nested Loops

```python
for i in range(1,5):
    for j in range(1,5):
        print(i, j)
```

```python
#Pattern printing
rows = int(input('Enter the rows: '))
for i in range(1,rows+1):
    for j in range(1, i+1):
        print("* ", end='')
    print()
    
```

```python
rows = int(input('Enter the rows: '))
for i in range(1,rows+1):
    for j in range(1, i+1):
        print(j, end='')
    print()
    for j in range(i-1, 0, -1):
        print(j, end='')
```

# Loop Control Statements

## Break

```python
# Break is used in Linear searching
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

```python
low = int(input('Enter the lower range: '))
up = int(input('Enter the upper range: '))

for i in range(low, up+1):
    for j in range(2, i):
        if i%j == 0:
            break
    else:
        print(i) 
```

## Continue

```python
for i in range(1, 10):
    if i == 5:
        continue
    print(i)
```

## Pass

```python
for i in range(1, 10):
    pass
```

## String Creation

```python
s = 'hello'
s = "hello"
# multiline strings
s = '''hello'''
s = """hello"""
s = str('hello')
print(s)
```

```python
s1 = "it's raining"
print(s1)
```

```python
## Accessing Substrings from a String
# Positive Indexing
s = 'hello world'
print(s[0])
```

```python
# Negative indexing
print(s[-1])
```

```python
# Slicing
print(s[0:5])  # 5 excluded
print(s[6:])
print(s[:7])
print(s[:])
```

```python
print(s[::2])
print(s[::3])
print(s[6::-1])
```

```python
print(s[::-2])
print(s[::-1])
```

```python
s = "hello world"
print(s[-5:])
print(s[-5::-1])
print(s[-1:-6])
```

## Editing or Deleting in Strings

```python
# Python Strings are immutable
s = 'hello world'
s[0] = 'H'
```

```python
s = 'hello world'
del s  # Delete whole strng
```

```python
s = 'hello world'
del s[-1:-5:2]    # Part of string cannot be deleted because of immutability
print(s)
```

```python
print('delhi'*5)
```

### 2. Relational Operators

```python
# Compare strings lexicographically on the basis of ascii value
print('delhi' == 'mumbai')
print('delhi' != 'mumbai')
print('delhi' >= 'mumbai')
print('delhi' <= 'mumbai')
print('delhi' > 'mumbai')
print('delhi' < 'mumbai')
```

### 6. isalnum/isalpha/isdigit/isidentifier

```python
print('anas1234'.isalnum())  #checks alpha-numeric characters in strings
print('anas123%'.isalnum())
print('anas'.isalnum())
```

```python
print('anas123'.isalpha())   #checks alphabets only
print('anas'.isalpha())
```

```python
print('123'.isdigit())    # checks only digits
print('anas'.isdigit())
print('anas123'.isdigit())
```

```python
print('first_name'.isidentifier())
print('first-name'.isidentifier())
```

### 7. Split/Join

```python
print('hi my name is Anas'.split())    # split all words in a string and stores in a list
print('Anas'.split())                  # breaks on space
print('hi my name is nitish'.split('i'))
print('hi my name is nitish'.split('is'))
```

```python
print("-".join(['hi', 'my', 'name', 'is', 'anas']))
print(" ".join(['hi', 'my', 'name', 'is', 'anas']))
```
