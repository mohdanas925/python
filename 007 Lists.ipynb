## Accessing Items from a List

### Indexing

```python
L = [1,2,3,4,5,[6,7]]
print(L[0])   # +ve indexing

print(L[-1])  # -ve indexing

print(L[5][0])
print(L[5][1])
```

```python
l = [[[1,2,3], [4,5,6], [7,8,9]]]
l1 = [[[1,2],[3,4]], [[5,6], [7,8]]]

print(l[0])
print(l[0][1])
print(l[0][2][1])

print(l1)
print(l1[0])
print(l1[1][0][0])
```

### Slicing

```python
l2 = [1,2,3,4,5,6]
print(l2[1:])
print(l2[-2:1:-1])
print(l2[::2])
```

## Adding Items to a List

### 1. Append

```python
# Adding only one single item to a list in the last index 
l3 = [1,2,3,4]
l3.append(True)
print(l3)
```

### 2. Extend

```python
l3.extend([7,8,5])
print(l3)
l3.extend('delhi')
print(l3)
```

### 3. Insert

```python
l4 = [1,2,3,4]
l4.insert(1,100)
print(l4)
```

## Editing items in a List

```python
l5 = [1,2,3,4,5]
l5[-1] = 600
print(l5)

l5[1:4] = 200, 300, 400
print(l5)
```

## Deleting items from a List

### 1. del

```python
l6 = [1,2,3]
print(l6)
del l6
print(l6)
```

```python
l6 = [1,2,3,4,5,6]
del l6[-1]
print(l6)

del l6[1:4]
print(l6)
```

### 3. Remove

```python
l7 = [1,2,3,4,5,6]
l7.remove(5)
print(l7)

l7.remove(9)
print(l7)
```

### 4. Pop

```python
a = [1,2,3,4,5]
print(a)
a.pop(0)   #pop(idx)
print(a)

a.pop()   #delete last element if index not given
print(a)
```

### 5. Clear

```python
a = [1,2,3,4,5]
print(a)

a.clear()
print(a)
```

## List Functions

```python
#min/max/len/sorted  (not permanent operations)
a8 = [1,2,3,5,6,3,6]

print(max(a8))
print(min(a8))
print(len(a8))
print(sorted(a8))
print(sorted(a8, reverse=True))
```

```python
# Count
z = [1,2,1,3,4,1,5]
z.count(1)
```

```python
# index
z.index(1)  #gives first index
```

```python
# reverse
z1 = [2,3,4,1,3,5,2]
# It permanently reverses the list
z1.reverse()
print(z1)
```

```python
# sort (permanently sort the list)
z2 = [2,1,5,4,5,2]
print(z2)
print(sorted(z2))
print(z2)

z2.sort()  #permanently sorted the list
print(z2)
```

```python
# copy -> shallow copy (a new copy of list is made on the another address with same elements)
z3 = [2,1,5,7,8]
print(z3)
print(id(z3))

z4 = z3.copy()
print(z4)
print(id(z4))
```

## List Comprehension

```python
# provides a conxise way of creating lists
# newlist = [expression for item in iterable if condition == True]

# Advantages
# 1. more time and space efficient than loops
# 2. requires fewer lines of codes
# 3. tramsforms iterative statement into a formula
```

```python
# Add 1 to 10 numbers in a list
l = []
for i in range(1,11):
    l.append(i)
print(l)
```

```python
l = [i for i in range(1,11,2)]
print(l)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3

[s*i for i in v]
```

```python
# Add squares
L = [1,2,3,4,5]
[i**2 for i in L]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5 == 0]
```

```python
languages = ['java', 'python', 'php', 'c', 'c++']
[lan for lan in languages if lan.startswith('p')]
```

```python
basket = ['apple', 'cherry', 'banana', 'guava']
my_fruit = ['apple', 'kiwi', 'grapes', 'banana']

# Add new list from my_fruit and items if the fruit exists in basket and also starts with 'a'
[fruit for fruit in my_fruit if fruit in basket if fruit.startswith('a')]
```

```python
# print 3x3 matrix using list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products on two lists
x1 = [1,2,3,4]
x2 = [5,6,7,8]

[i*j for i in x1 for j in x2]
```

### Zip function

```python
# Zip function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together
# and then the second item in each passed iterator are paired together
# If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator
```

```python
L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]
list(zip(L1, L2))
```

```python
# write a program to add items of 2 lists indexwise
[i+j for i, j in zip(L1, L2)]
```
