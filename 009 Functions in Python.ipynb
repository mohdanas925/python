## Let's create a function(with docstring)

```python
def is_even(num):   #When creating function(parameter)
    """
    This function returns if a given number is odd or even
    input - any valid integer
    output - odd/even
    created on - 16th Nov 2022
      """
    if type(num) == int:
        if num %2 == 0:
            return 'even'
        else:
            return 'odd'
    else:
        return 'Not Allowed'
```

```python
print(is_even.__doc__)
print(print.__doc__)
```

```python
for i in range(1, 9):
    x = is_even(i)     #when function call(argument)
    print(x)
```

```python
is_even('hello')
```

## Types of Arguments

### 1. Default Argument

```python
# Default arguments in Python are function parameters that are assigned a default value in the function definition. 
# If a value for that argument is not provided when the function is called, the default value is used. 
def power(a=1, b=2):
    return a**b
```

```python
power()
```

```python
power(2)
```

```python
power(2,3)
```

### 2. Positional Argument

```python
# Function parameter receives the input in the same order as they are sent in the arguments(function call)
power(3,3)
```

### 3. Keyword Argument

```python
# When the position of the parameters are not not known 
# The values passed to a function by explicitly specifying the parameter name along with its value
```

```python
power(b=3, a=2)
```

## *args and **kwargs

```python
# *args and **kwargs are special Python keywords that are used to pass the variable length of arguments to a function
# Use anything in place of (args)
```

```python
# *args
# allows us to pass a variable number of non-keyword arguments to a function.
# print is an example of *args

def multiply(*kwargs):
  product = 1

  for i in kwargs:
    product = product * i

  print(kwargs)
  return product
```

```python
multiply(1,2,3,4,5)
```

```python
# **kwargs
# **kwargs allows us to pass any number of keyword arguments.
# Keyword arguments mean that they contain a key-value pair, like a Python dictionary.

def display(**salman):

  for (key,value) in salman.items():
    print(key,'->',value)
```

```python
display(india='delhi', srilanka='colombo', nepal='kathmandu')
```

```python
# Points to remember while using *args and **kwargs
# order of the arguments matter(normal -> *args -> **kwargs)
# The words “args” and “kwargs” are only a convention, you can use any name of your choice
```

## Function without return statement

```python
# In Python, every function returns a value.
# A function that does not have an explicit return statement (often called a "void function" in other languages) will implicitly return the special value None. 
```

```python
def greet(name):
  print(f"Hello, {name}!")

# Call the function
greet("Alice")
# Output: Hello, Alice!
```

```python
def log_message(message):
  print(f"Log: {message}")

result = log_message("This is a message")
print(f"The return value is: {result}")
print(f"The type of the return value is: {type(result)}")

# Output:
# Log: This is a message
# The return value is: None
# The type of the return value is: <class 'NoneType'>
```

## Variable Scope

```python
def g(y):
    print(x)
    print(x+1)
x = 5
g(x)
print(x)
```

```python
def f(y):
    x = 1
    x += 1
    print(x)
x = 5
f(x)
print(x)
```

```python
# global variable does not change by the local variable
def h(y):  
    x += 1
x = 5
h(x)
print(x)
```

```python
def h(y):  
    global x 
    x += 1
x = 5
h(x)
print(x)
```

```python
def f(x):
   x = x + 1
   print('in f(x): x =', x)
   return x

x = 3
z = f(x)
print('in main program scope: z =', z)
print('in main program scope: x =', x)
```

## Nested Functions

```python
def f():
  def g():
    print('inside function g')
  g()
  print('inside function f')
```

```python
f()
```

```python
def g(x):
    def h():
        x = 'abc'
    x = x + 1
    print('in g(x): x =', x)
    h()
    return x

x = 3
z = g(x)
```

```python
def g(x):
    def h(x):
        x = x+1
        print("in h(x): x = ", x)
    x = x + 1
    print('in g(x): x = ', x)
    h(x)
    return x

x = 3
z = g(x)
print('in main program scope: x = ', x)
print('in main program scope: z = ', z)
```

## Lambda Function

```python
# A lambda function is a small anonymous function.
# A lambda function can take any number of arguments, but can only have one expression.

# Lambda a,b: a+b
#     (parameters):(expression)
```

```python
# x -> x^2
a = lambda x:x**2
a(2)
```

```python
# x,y -> x+y
a = lambda x,y:x+y
a(5,2)
```

### Difference between lambda vs Normal Function

```python
# No name
# Lambda has no return value(infact returns a function)
# not reusable
# They are used with HOF(Hogher Order Functions)
```

```python
# Check if a string has 'a'
a = lambda s:'a' in s
a('frdn')
```

```python
# odd or even
a = lambda x:'even' if x%2 == 0 else 'odd'
a(6)
```

### Higher Order Functions

```python
# A function which returns another function OR
# A function which has another function in input parameter
```

```python
def square(x):
    return x**2

# HOF
def transform(f, L):
    output = []
    for i in L:
        output.append(f(i))
    print(output)

L = [1,2,3,4,5]
transform(square, L)
transform(lambda x: x**3, L)
```

## Map

```python
# Square the items of a list
list(map(lambda x: x**2, [1,2,3,4]))
```

```python
# odd/even labelling of list items
L = [1,2,3,4,5]
list(map(lambda x: 'even' if x%2==0 else 'odd', L))
```

```python
# fetch names from a list of dict

users = [
    {
        'name':'Rahul',
        'age':45,
        'gender':'male'
    },
    {
        'name':'Nitish',
        'age':33,
        'gender':'male'
    },
    {
        'name':'Ankita',
        'age':50,
        'gender':'female'
    }
]

list(map(lambda user: user['name'], users))
```

## Filter

```python
l = [3,4,5,6,7]
list(filter(lambda x: x>=5, l))
```

```python
# Fetch fruits starting with 'a'
fruits = ['apple', 'mango', 'cherry', 'aqua']
list(filter(lambda x: x.startswith('a'), fruits))
```

## Reduce

```python
# Sum of all items
import functools

functools.reduce(lambda x,y: x+y, [1,2,3,4,5])
```

```python
# Find min
functools.reduce(lambda x,y: x if x<y else y, [23,11,4,45,10,6])
```
