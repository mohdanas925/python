### Write OOP classes to handle the following scenarios:

```python
# A user can create and view 2D coordinates
# A user can find out the distance between 2 coordinates
# A user can find find the distance of a coordinate from origin
# A user can check if a point lies on a given line
# A user can find the distance between a given 2D point and a given line
```

```python
class Point:

    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self):
        return '<{},{}>'.format(self.x_cod, self.y_cod)

    def euclidean_distance(self, other):
        return ((other.x_cod - self.x_cod)**2 + (other.y_cod - self.y_cod)**2)**0.5

    def distance_from_origin(self):
        return (self.x_cod**2 + self.y_cod**2)**0.5
        # return self.euclidean_distance(Point(0,0))


class Line:
    def __init__(self, A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return '{}x + {}y + {} = 0'.format(self.A, self.B, self.C)

    def point_on_line(line, point):
        if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
            return "Lies on the line"
        else:
            return "Does not lies on the line"


    def shortest_distance_from_line(line, point):
        return abs(line.A*point.x_cod + line.B*point.y_cod + line.C)/(line.A**2 + line.B**2)**0.5

    def is_lines_intersect(l1, l2):
        determinant = l1.A*l2.B - l2.A*l1.B
        if determinant != 0:
            return "Lines Intersect"
        else:
            return "Lines are Parallel or Coincident"
        
```

```python
p1 = Point(0,0)
p2 = Point(10,10)

# <x,y>
print(p1)
print(p2)

p1.euclidean_distance(p2)
```

```python
p2.distance_from_origin()
```

```python
l1 = Line(3,4,5)
print(l1)
```

```python
l1 = Line(1,1,-2)
p1 = Point(1,10)
print(l1)
print(p1)

l1.point_on_line(p1)
```

```python
l1.shortest_distance_from_line(p1)
```

```python
l1 = Line(1,2,1)
l2 = Line(2,3,5)
print(l1)
print(l2)

l1.is_lines_intersect(l2)
```

## How objects access attributes

```python
class Person:

    def __init__(self, name_in, country_in):
        self.name = name_in
        self.country = country_in

    def greet(self):
        if self.country == 'india':
            print('Namaste', self.name)
        else:
            print('Hello', self.name)
```

```python
# How to access attributes
p = Person('anas', 'india')
print(p.country)
print(p.name)
print(p.greet())
```

```python
# Attribute creation from outside of the class
p.gender = 'male'
p.gender
```

## Reference Variables

```python
# Reference variables hold the objects
# We can create objects without reference variable as well
# An object can have multiple reference variables
# Assigning a new reference variable to an existing object does not create a new object
```

```python
class Person:
    def __init__(self):
        self.name = 'nitish'
        self.gender = 'male'


p = Person()
# p is a reference variable which holds an object Person()
# p has a memory address of object
q = p
```

```python
print(id(p))
print(id(q))
```

```python
print(p.name)
print(q.name)
p.name = 'ankit'
print(p.name)
print(q.name)
```

## Pass by reference

```python
class Person:

    def __init__(self,name,gender):
        self.name = name
        self.gender = gender


# outside the class -> function
# function will takes an object as input attribute
# function will also returns an object
def greet(person):
    print('Hi my name is',person.name,'and I am a',person.gender)
    p1 = Person('ankit','male')
    return p1  
    

p = Person('anas', 'male')
x = greet(p)
print(x.name)
print(x.gender)
```

## Object mutability

```python
class Person:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class -> function
def greet(person):
  person.name = 'ankit'
  return person

p = Person('nitish','male')
print(id(p))
p1 = greet(p)
print(id(p1))

# objects are mutable
```

## Static Method

```python
# Static attributes are created at class level.
# Static attributes are accessed using ClassName.
# Static attributes are object independent. We can access them without creating instance (object) of the class in which they are defined.
# The value stored in static attribute is shared between all instances(objects) of the class in which the static attribute is defined.
```

```python
class Lion:
  __water_source="well in the circus"

  def __init__(self,name, gender):
      self.__name=name
      self.__gender=gender

  def drinks_water(self):
      print(self.__name,
      "drinks water from the",Lion.__water_source)

  @staticmethod
  def get_water_source():
      return Lion.__water_source

simba=Lion("Simba","Male")
simba.drinks_water()
print( "Water source of lions:",Lion.get_water_source())
```
