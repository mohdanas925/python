## Namespaces

```python
# A namespace is a space that holds names(identifiers).
# Programmatically speaking, namespaces are dictionary of identifiers(keys) and their objects(values)

# There are 4 types of namespaces:

# 1.Builtin Namespace
# 2.Global Namespace
# 3.Enclosing Namespace
# 4.Local Namespace
```

### Scope and LEGB Rule

```python
# A scope is a textual region of a Python program where a namespace is directly accessible.

# The interpreter searches for a name from the inside out, looking in the local, enclosing, global, and finally the built-in scope. 
# If the interpreter doesn’t find the name in any of these locations, then Python raises a NameError exception.
```

```python
# local and global
# global var
a = 2

def temp():
  # local var
  b = 3
  print(b)

temp()
print(a)
```

```python
# local and global -> same name
a = 2

def temp():
  # local var
  a = 3
  print(a)

temp()
print(a)
```

```python
# local and global -> local does not have but global has
a = 2

def temp():
  # local var
  print(a)

temp()
print(a)
```

```python
# local and global -> editing global
# we cannot perform operations on global varible from local space other than read
a = 2

def temp():
  # local var
  a += 1
  print(a)

temp()
print(a)
```

```python
a = 2

def temp():
  # local var
  global a
  a += 1
  print(a)

temp()
print(a)
```

```python
# local and global -> global created inside local
def temp():
  # local var
  global a
  a = 1
  print(a)

temp()
print(a)
```

```python
# local and global -> function parameters are local variables
def temp(z):
  # local var
  print(z)

a = 5
temp(5)
print(a)
# print(z)
```

```python
# Enclosing scope  LEGB rule
def outer():
  #inner/enclosed
  b = 3
  def inner():
    #local
    b = 4
    print(b)
  inner()
  print('outer function')

#global
b = 1
outer()
print('main program')
```

```python
# nonlocal keyword 
# changing non local variable from local space
def outer():
  a = 1
  def inner():
    nonlocal a
    a += 1
    print('inner',a)
  inner()
  print('outer',a)


outer()
print('main program')
```

## Decorators

```python
# A decorator in python is a function that receives another function as input and adds some functionality(decoration) to and it and returns it.

# This can happen only because python functions are 1st class citizens.

# There are 2 types of decorators available in python
# 1. Built in decorators like @staticmethod, @classmethod, @abstractmethod and @property etc
# 2. User defined decorators that we programmers can create according to our needs
```

```python
# simple example

def my_decorator(func):
    def wrapper():
        print("**************")
        func()
        print("**************")
    return wrapper

def hello():
    print("Hello")

def display():
    print("Hello Anas")

a = my_decorator(hello)
a()

b = my_decorator(display)
b()
```

```python
# closure
# A closure in Python is a nested function that remembers and can access variables from its enclosing scope, 
# even after the outer (enclosing) function has finished executing
def outer():
    a = 5
    def inner():
        print(a)
    return inner

b = outer()
b()
```

```python
# Better syntax
# Simple example
def my_decorator(func):
    def wrapper():
        print("**************")
        func()
        print("**************")
    return wrapper

@my_decorator
def hello():
    print("Hello")

hello()
```

```python
# meaningful example
import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        print('time taken by', func.__name__, time.time()-start,'secs')
    return wrapper


@timer
def hello():
    print('hello world')
    time.sleep(2)

@timer
def display():
    print('displaying something')
    time.sleep(4)

hello()
display()
```

```python
@timer
def square(num):
    time.sleep(1)
    print(num**2)

square(2)
```

```python
# meaningful example
# generic function which can take any number of inputs
import time

def timer(func):
    def wrapper(*args):
        start = time.time()
        func(*args)
        print('time taken by', func.__name__, time.time()-start,'secs')
    return wrapper

@timer
def square(num):
    time.sleep(1)
    print(num**2)

@timer
def power(a, b):
    print(a**b)
    
square(2)
power(2,3)
```

```python
def sanity_check(data_type):
  def outer_wrapper(func):
    def inner_wrapper(*args):
      if type(*args) == data_type:
        func(*args)
      else:
        raise TypeError('Ye datatype nai chalega')
    return inner_wrapper
  return outer_wrapper

@sanity_check(int)
def square(num):
  print(num**2)

@sanity_check(str)
def greet(name):
  print('hello',name)

square(2)
```
