## Class Relationship

```python
# 1. Aggregation (Has a relationship)
# Aggregation is a specialized form of association between two or more objects in which each object has its own life cycle but there exists an ownership as well.
# One class owns the other class
# Aggregation is typically a parent child relationship
# Example
# Customer has a address
# Restaurant has a menu
```

```python
# Private variable of owned class does not accessed by the owner class
# owner class
class Customer:

    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender
        self.address = address

    def print_address(self):
        print(self.address.get_city(), self.address.pin, self.address.state)
            #self.address._Address__city

    def edit_profile(self,new_name,new_city,new_pin,new_state):
        self.name = new_name
        self.address.edit_address(new_city,new_pin,new_state)

# owned class
class Address:

    def __init__(self, city, pin, state):
        self.__city = city
        self.pin = pin
        self.state = state

    def get_city(self):
        return self.__city

    def edit_address(self,new_city,new_pin,new_state):
        self.__city = new_city
        self.pin = new_pin
        self.state = new_state
    
```

```python
add1 = Address('gurgaon', 122011, 'haryana')
cust = Customer('anas', 'male', add1)

cust.print_address()
```

# Inheritance

```python
# Inheritance is a fundamental concept in object-oriented programming (OOP)
# Itallows a class (called a child or derived class) to inherit attributes and methods from another class (called a parent or base class)
```

```python
# parent
class User:

    def __init__(self):
        self.name = 'anas'
        self.gender = 'male'

    def login(self):
        print('login successful')

# child
class Student(User):

    # constructor
    # def __init__(self):
    #     self.rollnp = 100
    # if a constructor is present in the child class, 
    # then the parent class constructor will never call
    
    def enroll(self):
        print('enroll into the course')
```

```python
u = User()
s = Student()
print(s.name)
s.login()
s.enroll()
```

## Super

```python
# The super() function in Python makes class inheritance more manageable and extensible
# Super keyword used inside class, generally child class
# super cannot access variables
# super keyword is used to call the parent constructor when the child class also has a constructor
```

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        # syntax to call parent ka buy method
        super().buy()

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

```python
# super -> constuctor
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        print('Inside smartphone constructor')
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
        print ("Inside smartphone constructor")

s=SmartPhone(20000, "Samsung", 12, "Android", 2)

print(s.os)
print(s.brand)
```

```python
# can super access parent ka data?
# using super outside the class
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        # syntax to call parent ka buy method
        print(super().brand)

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

```python
# Does not use super outside the class
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        # syntax to call parent ka buy method
        # super().buy()

s=SmartPhone(20000, "Apple", 13)

s.super().buy()
```

```python
# Inheritance in summary

#A class can inherit from another class.
# Inheritance improves code reuse
# Constructor, attributes, methods get inherited to the child class
# The parent has no access to the child class
# Private properties of parent are not accessible directly in child class
# Child class can override the attributes or methods. This is called method overriding
# super() is an inbuilt function which is used to invoke the parent class methods and constructor
```

```python
class Parent:

    def __init__(self,num):
      self.__num=num

    def get_num(self):
      return self.__num

class Child(Parent):
  
    def __init__(self,num,val):
      super().__init__(num)
      self.__val=val

    def get_val(self):
      return self.__val
      
son=Child(100,200)
print(son.get_num())
print(son.get_val())
```

```python
# Multiple
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class Product:
    def review(self):
        print ("Customer review")

class SmartPhone(Phone, Product):
    pass

s=SmartPhone(20000, "Apple", 12)

s.buy()
s.review()
```

```python
# the diamond problem
# https://stackoverflow.com/questions/56361048/what-is-the-diamond-problem-in-python-and-why-its-not-appear-in-python2
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class Product:
    def buy(self):
        print ("Product buy method")

# Method resolution order
# When there is two parents of the single child class
# And having same method names in both parents
# Then the method will execute whose class appears first in the child class call 
class SmartPhone(Phone,Product):
    pass

s=SmartPhone(20000, "Apple", 12)

s.buy()
```

```python
# Method Overriding
# If there is inheritance and in both classes the methods are present with same name
# So, child's method will always executes
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

```python
# Method Overloading
# Method Overloading means defining multiple methods with the same name in the same class but with different parameters (different number or type of arguments).
# Python does not support method overloading
```

```python
class Shape:

    def area(self, radius):
        return 3.14* radius**2

    def area(self, l, b):
        return l*b
        
s = Shape()
print(s.area(2,3))
print(s.area(2))
```

```python
class Shape:

    def area(self, a,b=0):
        if b == 0:
            return 3.14*a*a
        else:
            return a*b


s = Shape()
print(s.area(2))
print(s.area(2,3))
```

# Abstraction

```python
# Abstraction means hiding implementation details and showing only the essential features of an object.
# In simple words:
# Show what an object does, not how it does it.

# When you use an ATM machine:
# You press buttons to withdraw money.
# You don’t know how the internal banking system works.
# That is abstraction.
```

```python
from abc import ABC,abstractmethod
class BankApp(ABC):

  def database(self):
    print('connected to database')

  @abstractmethod
  def security(self):
    pass

  @abstractmethod
  def display(self):
    pass
```

```python
class MobileApp(BankApp):

  def mobile_login(self):
    print('login into mobile')

  def security(self):
    print('mobile security')

  def display(self):
    print('display')
```

```python
mob = MobileApp()
```

```python
mob.security()
```
