# 1. Operators

```python
# Arithmetric Operators
print(5+6)

print(5-6)

print(5*6)

print(5/2)

print(5//2)

print(5%2)

print(5**2)
```

```python
# Relational Operators
print(4>5)

print(4<5)

print(4>=4)

print(4<=4)

print(4==4)

print(4!=4)
```

```python
# Logical Operators
print(1 and 0)

print(1 or 0)

print(not 1)
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

print(~3)

print(4 >> 2)

print(5 << 2)
```

```python
# Assignment Operators

# = 
# a = 2

a = 2

# a = a % 2
a %= 2

# a++ ++a

print(a)
```

```python
# Membership Operators

# in/not in

print('D' not in 'Delhi')

print(1 in [2,3,4,5,6])
```

```python
# Program - Find the sum of a 3 digit number entered by the user

number = int(input("Enter the 3 digit number: "))
a = number%10
number = number//10

b = number%10
number = number//10

c = number%10

print(a+b+c)
```

# 2. If-else

```python
username = input("Enter the username: ")
password = int(input('Enter the password'))

if username == 'anas' and password == 1234:
    print('Login Successfully! ')
elif username == 'anas' and password != 1234:
    print("Wrong password")

    password = int(input("Enter Password Again! "))
    if password == 1234:
        print("Finally Login Successfully! ")
    else:
        print('Wrong Password, Tumse n ho paega beta! ')
else:
    print('Wrong Username or Password! ')
```

# Loops

```python
number = int(input('enter the number'))

i = 1

while i<11:
  print(number,'*',i,'=',number * i)
  i += 1
```

```python
# while loop with else 

x = 1

while x < 3:
  print(x)
  x += 1

else:
  print('limit crossed')
```

```python
# Guessing Game
num = random.randint(1, 100)
guess = int(input('Enter a number: '))
count = 1

while (guess != num):
    if(guess < num):
        print('Guess higher! ')
    else:
        print('Guess lower! ')
    count += 1
    guess = int(input("Enter! "))
else:
    print('Correct guess in times = ', count)
```

```python
# Sequence sum 1/1! + 2/2! + 3/3! ....
n = int(input('Enter the number: '))

result = 0
fact = 1
for i in range(1, n+1):
    fact = fact*i
    result += i/fact

print(result)
```
