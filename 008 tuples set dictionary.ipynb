# Tuples

```python
# A tuple in python is similar to a list. 
# The difference between the two is that we cannot change the elelments of a tuple once it is assigned whereas we can change the elements of list
# Tuple is an immutable
```

## Characteristics

```python
# Ordered
# Unchangable
# Allows duplicate elements
```

## Tuples Creation

```python
# Empty tuple
t1 = ()
print(t1)

# Tuple with a single item
t2 = (1,)
print(t2)
print(type(t2))

# homogenous tuple
t3 = (1,2,3,4,5)
print(t3)

# heterogenous tuple
t4 = (1, 1.4, True, 'Anas')
print(t4)
```

```python
# 2D tuple
t5 = (1,2,3,4,(5,6))
print(t5)

# Using type conversion
t6 = tuple('hello')
print(t6)
```

## Accessing Elements

### 1. Indexing

```python
print(t3)
print(t3[0])
print(t3[-1])
```

### 2. Slicing

```python
print(t3[::-1])
```

```python
print(t5)
print(t5[4][0])

# t5[smaller : greater : +ve]
# t4[greater : smaller : -ve]
```

## Editing elements in a tuple

```python
# tuples are immutable so they wil never be edited
print(t3)
t3[0] = 100
```

## Adding Items

```python
# Tuples are immutable so no element in a tuple would be added
```

## Deletion in Tuple

```python
# Single item or portion from a tuple will never be deleted
print(t3)
del t3[-1]
```

```python
# Whole tuple would be deleted 
print(t3)
del t3
print(t3)
```

## Operations on Tuples

```python
# Add
x1 = (1,2,3,4)
x2 = (5,6,7,8)

print(x1 + x2)
```

```python
# Multiplication
print(x1*3)
```

```python
# Membership operator
print(1 in x1)

print(1 not in x1)
```

```python
# Iteration
for i in x1:
    print(i)
```

## Tuple functions

```python
# len/sum/min/max/sorted

print(len(x1))
print(sum(x1))
print(min(x1))
print(max(x1))
print(sorted(x1))
print(sorted(x1, reverse = True))
```

### 1. Count

```python
# Tuple specific
t = (1,5,3,5,2,5,2)
print(t.count(5))
print(t.count(4))
```

### 2. Index

```python
print(t.index(5))      # gives first occurence of element
print(t.index(50))
```

## Difference betweeen lists and tuples

```python
# Syntax
# Mutability
# mutable datatypes are faster
# mutable data types takes less memory
# Lists are more error prone
```

```python
import time
L = list(range(10000000))
T = tuple(range(10000000))

start = time.time()
for i in L:
    i*5
print('List time: ', time.time()-start)

start = time.time()
for i in T:
    i*5
print('Tuple time ', time.time()-start)
```

```python
import sys 
L = list(range(10000))
T = tuple(range(10000))

print('List size:', sys.getsizeof(L))
print('Tuple size:', sys.getsizeof(T))
```

# Sets

```python
# A set is an unordered collection of items.
# Every set element is unique (no duplicate) and must be immutable
# However a set itself is mutable. We can add or remove items from it.
# Can used to perform mathematical operations like nion, intersection, symmetric diff rtc.
```

## Characteristics

```python
# Unordered
# Mutable
# No duplicate (Unique)
# Can't contain mutable data types
```

## Creating Sets

```python
# Empty set
s = set()
print(s)
print(type(s))

# 1D Set
s1 = {1,2,3}
print(s1)

# 2D Set
# s2 = {1,2,3, {4,5}}
# 2D sets are not allowed in set

# Heterogenous
s3 = {1, 1.5, 'hello', True, (1,2,3)}  # True = 1
print(s3)

# Using type conversion
s4 = set([1,2,3])
print(s4)
```

```python
# Duplicates not allowed
s5 = {1,1,2,3,4,4,5,3}
print(s5)
```

```python
# Set cannot have mutable items
s6 = {1,2,5,[3,4]}
print(s6)
```

## Accessing Elements

```python
# Accessing items, indexing and slicing in a set is not allowed because sets are unordered
```

## Editing Items

```python
# Editing new values in place of old values in a set is not allowed
```

### 1. Adding Items

```python
# Add
s = {1,2,3,4}
s.add(5)
print(s)
```

```python
# Update
print(s)
s.update([6,7,8])
print(s)
```

### 2. Deleting Items

```python
# del
s = {1,2,3,4,5}
print(s)
del s  # Whole set is deleted

# del s[0] Particular item from a set will not be deleted
```

```python
# Discard
# Discard will not through an error if the element deleting is not present in a set
s = {1,2,3,4,5}
print(s)
s.discard(5)
print(s)

s.discard(50)   # item not present and gives out no error
print(s)
```

```python
# Remove
# Remove will through an error if the element deleting is not present in a set
print(s)
s.remove(4)
print(s)

s.remove(50)
print(s)
```

```python
# pop  
# delete a random element from a set
print(s)
s.pop()
print(s)
```

```python
# Clear
# Clear whole set
print(s)
s.clear()
print(s)
```

## Set Operations

```python
# Union(|)
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1 | s2
```

```python
# Intersection(&)
s1 & s2
```

```python
# Difference(-)
print(s1 - s2)
print(s2 - s1)
```

```python
# Symmetric Difference(^)
# Printing all except common elements
print(s1 ^ s2)
print(s2 ^ s1)
```

```python
# Membership Test
print(1 in s1)
print(1 not in s1)
```

```python
# Iteration
for i in s1:
    print(i)
```

## Set Functions

```python
# len/sum/min/max/sorted
s = {3,1,4,5,2,7}
print(sum(s))
print(len(s))
print(min(s))
print(max(s))
print(sorted(s))
print(sorted(s, reverse=True))
```

```python
# Union/ update
# permanent change in a set
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.update(s2)

print(s1)
print(s2)
```

```python
# Intersection / intersection_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.intersection_update(s2)

print(s1)
print(s2)
```

```python
# Difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.difference_update(s2)

print(s1)
print(s2)
```

```python
# Symmetric_difference
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.symmetric_difference_update(s2)

print(s1)
print(s2)
```

```python
# isdisjoint
# All elements in the sets are unique(distinct sets)
s1 = {1,2,3,4}
s2 = {4,5,5,6}

s1.isdisjoint(s2)
```

```python
# issubset
s1 = {1,2,3,4,5}
s2 = {3,4,5}

print(s2.issubset(s1))
print(s1.issubset(s2))
```

```python
# issuperset
s1 = {1,2,3,4,5}
s2 = {3,4,5}

print(s2.issuperset(s1))
print(s1.issuperset(s2))
```

```python
# copy
# makes a shallow copy on different location
s1 = {1,2,3}
s2 = s1.copy()
```

# Dictionary

```python
# Dictionary in python is a collection of key values, used to store data values like a map, which, unlike other data types which hold only a single value as an element.
# In some languages it is known as map or assosiative arrays.
# dict = { 'name' : 'nitish' , 'age' : 33 , 'gender' : 'male' }
```

## Characteristics

```python
# Mutable
# Unordered
# Indexing has no meaning
# keys can't be duplicated
# keys can't be mutable items
```

## Create Dictionary

```python
# Empty
d = {}
print(d)

# 1D dictionary
d1 = { 'name' : 'nitish' ,'gender' : 'male' }
print(d1)

# with mixed keys
d2 = {(1,2,3):1,'hello':'world'}
print(d2)

# 2D dictionary -> JSON
s = {
    'name':'nitish',
     'college':'bit',
     'sem':4,
     'subjects':{
         'dsa':50,
         'maths':67,
         'english':34
     }
}
print(s)
```

```python
# using sequence and dict function
d4 = dict([('name','nitish'),('age',32),(3,3)])
print(d4)

# duplicate keys not allowed
d5 = {'name':'nitish','name':'rahul'}
print(d5)

# mutable items as keys not allowed (list)
d6 = {'name':'nitish',(1,2,3):2}
print(d6)
```

## Accessing Elements

```python
my_dict = {'name': 'Jack', 'age': 26}
# []
print(my_dict['age'])

# get
print(my_dict.get('age'))

s['subjects']['maths']
```

## Adding New Key Value Pair

```python
d4['gender'] = 'male'
print(d4)
d4['weight'] = 72
print(d4)

s['subjects']['ds'] = 75
print(s)
```

## Removing Key-Value Pairs

```python
d = {'name': 'nitish', 'age': 32, 3: 3, 'gender': 'male', 'weight': 72}
# pop
d.pop(3)
print(d)
```

```python
# popitem  (Removes the last key value pair)
d.popitem()
d.popitem()
print(d)
```

```python
# clear  (clars all key value pairs)
d.clear()
print(d) 
```

```python
# del
del d['name']
print(d)

del s['subjects']['maths']
s
```

## Editing Key Value Pairs

```python
print(s)
s['subjects']['dsa'] = 80
print(s)
```

## Dictionary Operations

```python
# Membership
print(s)
print('name' in s)
```

```python
# Iteration
dict = {'name': 'anas', 'gender':'male', 'age': 22}
for i in dict:
    print(i, dict[i])
```

```python
# len/sorted/min/max
print(len(dict))
print(dict)

# Sorted
print(sorted(dict))
print(sorted(dict, reverse=True))

# min key
print(min(dict))

# max key
print(max(dict))
```

```python
# items/keys/values
print(dict)

#items
print(dict.items())

# keys
print(dict.keys())

# values
print(dict.values())
```

```python
# update (update a given dictionary with the other dictionary)
d1 = {1:2, 3:4, 4:5}
d2 = {4:7, 6:8}
print(d1)
print(d2)


d1.update(d2)
print(d1)
```

## Dictionary Comprehension

### {key: value for vars in iterable }

```python
# print 1st 10 numbers and their squares
{i:i**2 for i in range(1,11)}
```

```python
distances = {'delhi':1000,'mumbai':2000,'bangalore':3000}
print(distances.items())
```

```python
# Distances km to miles
distances = {'delhi':1000,'mumbai':2000,'bangalore':3000}
{key: value*0.62 for (key, value) in distances.items()}
```

```python
# using zip
days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

{key:value for (key,value) in zip(days, temp_C)}
```

```python
# using if condition
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

{ key:value for (key,value) in products.items() if value>0}
```

```python
# Nested Comprehension
# print tables of number from 2 to 4

{i: {j:i*j for j in range(1,11)} for i in range(2,5)}
```
